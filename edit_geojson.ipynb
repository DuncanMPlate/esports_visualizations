```python
import geopandas as gp
import os
import pandas as pd
```

```python
## Get our two dataframes
year_subregions_path = os.path.join(os.getcwd(), "Resources", "Year_Subregions_Clean.csv")
geo_json_path = os.path.join(os.getcwd(), "Resources", "map_test.geojson")
year_subregions_df = pd.read_csv(year_subregions_path)
geo_json_df = gp.read_file(geo_json_path)
geo_json_df.head()
```

```python
geo_json_df = geo_json_df.drop(geo_json_df.columns[4], axis=1)
geo_json_df.head()
```

```python
year_subregions_df = year_subregions_df.rename(columns = {"Region" : "NAME"})
year_subregions_df.head()
```

```python
## Now we can merge
final_geojson = geo_json_df.merge(year_subregions_df, on='NAME')
final_geojson.head()
```

```python
final_geojson.to_file('Resources/map_trends.geojson', driver='GeoJSON') 
```
