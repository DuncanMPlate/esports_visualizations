```python
import pandas as pd
```

```python
years = [2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021]
```

```python
base_df = pd.read_csv('Resources/year_subregions/2006.csv')
base_df
```

```python

for year in years:
    next_df = pd.read_csv(f'Resources/year_subregions/{year}.csv')
    base_df = base_df.join(next_df.set_index('Region'), on = 'Region')
    
```

```python
base_df = base_df.fillna('0%')
```

```python
base_df = base_df.replace({'<':''}, regex=True)
base_df = base_df.replace({'%':''}, regex=True)
```

```python
base_df
```

```python
base_df.to_csv('Resources/Year_Subregions_Clean.csv', index=False)
```
